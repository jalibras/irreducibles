```python
from sparsity import MyGraph
```

```python
vertex_parts = [[1,2,7],[3,4,8],[5,6]]
edge_parts = [[1,3],[2,5],[4,6],[7,8]]
g = MyGraph(dart_partitions=[vertex_parts,edge_parts])
```

```python
g.edges()
```

```python
g.plot()
```

```python
from sparsity import PebbleGame
```

```python
p = PebbleGame(g.vertices(),2,2) # initialise a Pebble to test for (2,2)-sparsity. We must specify the vertex set of the graph
```

```python
p.run(g) # tests g for sparsity. If g is sparse it will return the graph object. If it is not it will return False.
```

```python
from rotation import OrientedRotationSystem
```

```python
l = OrientedRotationSystem.from_mygraph(g) # find all irreducible embeddings of g in the torus
```

```python
l
```

```python
ors = l[0]
```

```python
ors.faces()
```
